```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
HousingData_A3 = pd.read_csv("HousingData_A3.csv")
HousingData_A3.drop(['Unnamed: 0'],axis=1,inplace=True)
HousingData_A3.drop(['ID'],axis=1,inplace=True)
```

# Removing the multicollinearity correlation between IV's

```python
HousingData_A3.head()
```

```python
HousingData_A3['Waterfront_View_Yes'].head()
```

```python
# Scaling The Data
from sklearn.preprocessing import StandardScaler as SS
datascale = SS()
Y = HousingData_A3['Sale_Price']
X = datascale.fit_transform(HousingData_A3.drop(columns=['Sale_Price']))
# for data purpose creating dataframe
X = pd.DataFrame( X,columns = HousingData_A3.drop(columns=['Sale_Price']).columns)
X.head(3)
```

```python
X.corr()
```

```python
# to find higher correlation among the independant variables 
# having correlation greater than 0.5
correlationIV = X.corr()
# forming list of pairs of independant variables having correlation greater than 0.5 tkaing absolute value
for_pairs = list()
for i in correlationIV.columns:
    for j in correlationIV.columns:
        if(correlationIV.loc[i,j]>abs(0.5) and (i!=j)):
            
            for_pairs.append([str(i),str(j)])
len(for_pairs)
for_pairs
# To remove the independant variables having multicollinearity vif is used variance inflation factor
```

```python
# for this using statsmodel
# the above will have some discrepancies due to repeatation of attributes
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = X
listofvifvalues = [variance_inflation_factor(vif.values,i) for i in range(vif.shape[1])]
listofvifvaluesseries = pd.Series(listofvifvalues,index = vif.columns)
listofvifvaluesseries
# its shows only 7 variables have vif value greater than 5 and that should be removed
# now for calculating nad minimising the vif values iterativ process can be used
```

```python
# removing the columns 
def remove_columns(data):
    tmp_list = [variance_inflation_factor(data.values,i) for i in range(data.shape[1])]
    vif_tmp = pd.Series(tmp_list,data.columns)
    if(vif_tmp.max()>5):
        data = data.drop(columns = [vif_tmp[vif_tmp==vif_tmp.max()].index[0]])
        return data
    else:
        return data
```

```python
for i in range(7):
    vif = remove_columns(vif)
vif.head()
```

# FINAL VALUES HAVING VIF < 5  

```python
finalvif = pd.Series([variance_inflation_factor(vif.values,i) for i in range(vif.shape[1])],index = vif.columns)
finalvif
```

# Linear Regression

```python
# taking independant variables after scaling them from standard scalar
x= vif
y = HousingData_A3['Sale_Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 101)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression(normalize=True)
# makes b0 coefficinet zero
LR.fit(x_train,y_train)
# it performs the gradient descent algorithm
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
len(LR.coef_)
LR.coef_
# set of parameters m1,m2,m3...
```

```python
# testing generating predictions 
predictions = LR.predict(x_test)
predictions
```

```python
# calculating the r2 value without the sklearn
ymanual = pd.DataFrame({'Salep':y_test})
ymanual['mean_sales'] = y_test.mean()
from sklearn.metrics import mean_squared_error as mse
mse_model = mse(predictions,y_test)
mse_base = mse(ymanual['mean_sales'],ymanual['Salep'])
r2 = 1-(mse_model/mse_base)
r2
```

```python
# calculating the r2 value with the sklearn
LR.score(x_test,y_test)
```

```python
result = pd.DataFrame({'DataCols':x_train.columns,'Coefficients':LR.coef_})
# result
plt.figure(figsize=(12,10))
plt.barh(result.DataCols,result.Coefficients,color='grey')
plt.xlabel('Coefficients')
plt.ylabel('Variables')
plt.show()
```
