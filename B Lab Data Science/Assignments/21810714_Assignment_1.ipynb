# Converting variable data type 

```python
mapping = {"None": '0',
        "Once": '1',
        "Twice": '2',
        "Thrice": '3',
        "Four": '4'}
HousingData["No of Times Visited"] = HousingData["No of Times Visited"].map(mapping)
```

```python
#Initially the datatypr of No of Times visited column is Object
HousingData['No of Times Visited'].dtype
```

```python
#Converting it to integer
HousingData['No of Times Visited'] = HousingData['No of Times Visited'].astype(int)
```

```python
HousingData.info()
#data type has been changed to integer from object
```

# Missing values,Filling them and treating outliers

```python
#There are some outliers that are needed to be removed 
#1230000 is an outlier
HousingData['Sale Price'].head(10)
```

```python
#visualising with box plot
import seaborn as sb
sb.boxplot(HousingData['Sale Price'])
```

```python
HousingData['Sale Price'].describe()
```

```python
q1 = HousingData['Sale Price'].quantile(0.25)
q3 = HousingData['Sale Price'].quantile(0.75)

iqr = q3 - q1
lowerbound = q1 - 1.5*iqr
upperbound = q3 + 1.5*iqr

def imputation(value):
    if(value>upperbound):
        return upperbound
    elif(value<lowerbound):
        return lowerbound
    else:
        return value

HousingData['Sale Price'] = HousingData['Sale Price'].apply(imputation)
HousingData['Sale Price'].describe()
```

```python
HousingData.info()
```

```python
HousingData.describe()
```

```python
#imputing method is not considered for filling the missing values of target variable
#using deletion method for treating missing values
#on target variable
HousingData.dropna(inplace = True,axis = 0 ,subset=['Sale Price'])
```

```python
sum_of_null = HousingData.isnull().sum()
sum_of_null[sum_of_null!=0]
```

# Treating remaining missing values of independant variable using simpleimputer

```python
numerical_column = ['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)',
                    'Area of the House from Basement (in Sqft)',
                    'Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
imputer = SimpleImputer(missing_values = np.nan,strategy = 'median')
HousingData[numerical_column] = imputer.fit_transform(HousingData[numerical_column])
```

```python
sum_of_null = HousingData.isnull().sum()
sum_of_null[sum_of_null!=0]
#zipcode should be treated as categorical variable
```

```python
#reshaping the zipcode column as impute.fit takes today array
column_zipcode = HousingData['Zipcode'].values.reshape(-1,1)
```

```python
imputer_2 = SimpleImputer(missing_values = np.nan,strategy = 'most_frequent')
HousingData['Zipcode'] = imputer_2.fit_transform(column_zipcode)
```

```python
sum_of_null = HousingData.isnull().sum()
sum_of_null[sum_of_null!=0]
#finally no missing values 
```

```python
#converting integer to object for zipcode attribute
HousingData['Zipcode'] =HousingData['Zipcode'].astype(object)  
```
