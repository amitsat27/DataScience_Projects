```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression(normalize=True)
# makes b0 coefficinet zero
LR.fit(X_train,Y_train_up)
```

```python
# testing generating predictions 
predictions = LR.predict(X_test)
predictions
```

```python
# calculating predictions
residuals = predictions - Y_test_up
residual_table = pd.DataFrame({'Residuals':residuals,'Predictions':predictions})
```

```python
Y_test_up
predictions
```

```python
# since there are multiple independant variables
residual_table = residual_table.sort_values(by='Predictions')
```

```python
# plotting a single line at 0 point to get distribution of data 
# line = [i for i in range()]
linex = [i for i in range(int(residual_table['Predictions'].max()))]
liney = [0 for i in range(int(residual_table['Predictions'].max()))]
```

```python
plt.scatter(residual_table['Predictions'],residual_table['Residuals'],s=2)
plt.plot(linex,liney,color='red',label='Line')
plt.ylim(-400000,400000)
plt.xlabel('Residuals')
plt.ylabel('Fitted Points')
plt.title('Residual Plot')
plt.show()
```

```python
# Residual plot
plt.hist(residual_table['Residuals'], color = 'green', bins = 200)
plt.xlabel('Residuals')
plt.ylabel('frequency')
plt.title('Distribution of Residuals')
plt.show()
```

```python

x = datachurnafter.drop(columns = ['churn','customer_id'],axis = 1)
y = datachurnafter['churn']
```

```python
from sklearn.preprocessing import StandardScaler as SS
scalar = SS()
x = scalar.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train , y_test =  tts(x,y,train_size = 0.8,random_state=20,stratify=y)
```

```python
x_train.shape, x_test.shape, y_train.shape , y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
splitscale = SS()
splitscale.fit_transform(x_train,x_test)
```

```python
# instance of logisticregression
logisticclassifier = LogisticRegression(class_weight='balanced')
# fitting the data
logisticclassifier.fit(x_train,y_train)
```

```python
PredictionsLRchurn = logisticclassifier.predict(x_test)
PredictionsLRchurn
```

```python
from sklearn.metrics import classification_report,accuracy_score
print("For Logistic Regression")
print('Accuracy Score :',accuracy_score(y_test,PredictionsLRchurn))
print( classification_report(y_test,PredictionsLRchurn))
```

```python
predictedprobalities = logisticclassifier.predict_proba(x_test)
```

```python
predictedprobalities[:,1]
```

# precision recall curve

```python
from sklearn.metrics import precision_recall_curve
precpoints,recallpoints,thresholdpoints = precision_recall_curve(y_test,predictedprobalities[:,1])
```

```python
plt.plot(thresholdpoints,precpoints[:-1],label='Precision')
plt.plot(thresholdpoints,recallpoints[:-1],label='Recall')
plt.xlabel("Threshold Points")
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
Coeffs = logisticclassifier.coef_.reshape(-1)
```

```python
x = datachurnafter.drop(columns = ['churn','customer_id'],axis = 1)
```

```python
# AUC ROC CURVE
from sklearn.metrics import roc_curve,roc_auc_score
for i in range(2):
    FPR[i],TPR[i],thresholds = roc_curve(y_test,predictedprobalities[:,1])
```

```python
plt.plot(FPR,TPR,color='green')
plt.plot([0,1],[0,1],color='red')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('AUC-ROC Curve')
plt.show()
roc_auc_score(y_test,predictedprobalities[:,1])
```
